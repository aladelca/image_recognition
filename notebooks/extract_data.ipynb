```python
import pandas as pd
from google_images_search import GoogleImagesSearch
import os
os.chdir('..')
```

```python

developer_key = os.getenv('GCS_DEVELOPER_KEY')
cx = os.getenv('GCS_CX')
```

```python
gis = GoogleImagesSearch(developer_key, cx)
query = 'Neymar'
_search_params = {
    'q': query,
    'num': 200,
    'fileType': 'jpg',
    'rights': 'cc_publicdomain',
    #'safe': 'active|high|medium|off|safeUndefined', ##
    #'imgType': 'clipart|face|lineart|stock|photo|animated|imgTypeUndefined', ##
    #'imgSize': 'huge|icon|large|medium|small|xlarge|xxlarge|imgSizeUndefined', ##
    #'imgSize': 'large',
    #'imgType': 'face',
    #'imgDominantColor': 'black|blue|brown|gray|green|orange|pink|purple|red|teal|white|yellow|imgDominantColorUndefined', ##
    #'imgColorType': 'color' ##
}

gis.search(search_params=_search_params, path_to_dir=f'data/{query}', width=500, height=500)
```

```python
from icrawler.builtin import GoogleImageCrawler, BingImageCrawler
query = "robert lewandowski"
google_Crawler = GoogleImageCrawler(storage = {'root_dir': f'data/crawler/{query}'})
google_Crawler.crawl(keyword = query, max_num = 800)

bing_Crawler = BingImageCrawler(storage = {'root_dir': f'data/crawler/{query}_bing'}) 
bing_Crawler.crawl(keyword = query, max_num = 800)
```
